```python
# Soal IPA dan penilaian

# Daftar soal
soal = [
    {
        "pertanyaan": "1. Apa nama planet terbesar di tata surya?",
        "pilihan": ["a. Mars", "b. Bumi", "c. Jupiter", "d. Venus"],
        "jawaban": "c"
    },
    {
        "pertanyaan": "2. Air berubah menjadi uap air melalui proses apa?",
        "pilihan": ["a. Kondensasi", "b. Evaporasi", "c. Presipitasi", "d. Sublimasi"],
        "jawaban": "b"
    }
]

skor = 0

# Looping soal
for item in soal:
    print(item["pertanyaan"])
    for pilihan in item["pilihan"]:
        print(pilihan)
    jawaban_user = input("Jawaban Anda (a/b/c/d): ").lower()
    if jawaban_user == item["jawaban"]:
        print("Benar!\n")
        skor += 1
    else:
        print(f"Salah! Jawaban yang benar adalah: {item['jawaban']}\n")

# Hitung nilai
nilai = skor / len(soal) * 100

# Tentukan grade
if nilai == 100:
    grade = "A"
elif nilai >= 80:
    grade = "B"
elif nilai >= 60:
    grade = "C"
else:
    grade = "D"

# Tampilkan hasil
print(f"Skor Anda: {skor} dari {len(soal)} soal")
print(f"Nilai: {nilai}")
print(f"Grade: {grade}")
```
